```python
""" My attempt to make a particle move across the screen """

# Standard preamble
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
% matplotlib notebook

# Initial conditions
x, y = 0, 0  # initial location for point
vx1, vy1 = 5, 0  # velocity values for the point in each axis direction
t = 0  # initial value for time that just serves as a counter

# Define the function for point 1
def p1(t):
    return np.array([x + vx1 * t, y + vy1 * t])

# Plots the figure
fig, ax = plt.subplots()
ax.axis([0, 500, -200, 200])
ax.set_aspect('equal')
point, = ax.plot(0, 1, marker='o', markersize=8)

# Updates the points as time (t) increases
def updatefig(t):
    t += 1
    x, y = p1(t)
    point.set_data([x], [y])
    return point,

# Animates the updating figure
ani = animation.FuncAnimation(fig, updatefig, interval=100, blit=True)
```
